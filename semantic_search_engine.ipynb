**Document object has page_content , metadata , id**

```python
from langchain_core.documents import Document

documents = [
    Document(page_content="A bunch of scientists bring back dinosaurs and mayhem breaks loose"),
    Document(page_content="Leo DiCaprio gets lost in a dream within a dream within a dream within a dream"),
    Document(page_content="A psychologist / detective gets lost in a series of dreams within dreams within dreams and dreams")

]
```

# **Document Loader**

```python
from langchain_community.document_loaders import PyPDFLoader
file_path = "/content/drive/MyDrive/pdfs/Sandeep Singh Resume improved version.pdf"
loader = PyPDFLoader(file_path)
docs = loader.load()
print(len(docs))
```

# **PyPDFLoader makes a list Document object per page of pdf**

```python
print(f"{docs[0].page_content}")
print("------------------------------")
print(f"{docs[0].metadata}")
```

# **Text Splitter**
When using a document splitter (like CharacterTextSplitter) in LangChain, setting **add_start_index=True** ensures that each chunk (or split document) retains information about where it started in the original document.

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 100,
    chunk_overlap = 20,
    add_start_index = True
)

all_splits = text_splitter.split_documents(docs)
print(len(all_splits))
```

```python
#To check if overlapping is actually occuring or not
for i in all_splits:
  print(i.page_content)
  print(i.metadata["start_index"])
  print("______________")
```

# **Vector Embeddings**
we can use different vector providers(like google , aws , microsoft , nvidia) ,
 here i m using HuggingFace (cause its free)

```python
!pip install -qU langchain-huggingface
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
```

```python
vector_1 = embeddings.embed_query(all_splits[0].page_content)
vector_2 = embeddings.embed_query(all_splits[1].page_content)

assert len(vector_1) == len(vector_2)

```

```python
print(len(vector_1))
print(len(vector_2))
print(vector_1[0:10])
```

## **Vector stores object stores document as vectors and are used for efficient similarity search**

it stores the embedding model we are using to vectorize and stores are text or document

```python
!pip install -qU langchain-core
```

```python
from langchain_core.vectorstores import InMemoryVectorStore

vector_store = InMemoryVectorStore(embeddings)
```

You’re adding a list of documents (all_splits) into a vector store (vector_store), and it returns their unique IDs.

```python
ids = vector_store.add_documents(documents=all_splits)
```

```python
print(ids)
```

# **similarity and maximum marginal relevance (to balance similarity with query to diversity in retrieved results)**

# **Usage / Result**

```python
results = vector_store.similarity_search(
   # "what are achievements?"
   "technology"
)

print(type(results))
print(len(results))

for i in results:
  print(i.metadata)
  print("______________________________________")

for i in results:
  print(i.page_content)
  print("-------------------------------------")
```

similarity_search() is synchronous and can lead to blockings.

so we have asimilarity_search()

| Feature            | `similarity_search()`     | `asimilarity_search()`                               |
| ------------------ | ------------------------- | ---------------------------------------------------- |
| **Sync or Async?** | Synchronous (blocking)    | Asynchronous (non-blocking)                          |
| **When to use?**   | Normal Python scripts     | Inside `async def` functions or async apps           |
| **Why use async?** | Not needed for small apps | Needed for high concurrency (e.g. FastAPI, web apps) |


```python
results = await vector_store.asimilarity_search("phone number")

for i in results:
  print(i.metadata)
  print("______________________________________")

for i in results:
  print(i.page_content)
  print("-------------------------------------")
```

```python
# Note that providers implement different scores; the score here
# is a distance metric that varies inversely with similarity.
results = vector_store.similarity_search_with_score("what are the soft skills")
doc, score = results[0]
print(f"Score: {score}\n")
print(doc.page_content)
```

```python
queryEmbedding = embeddings.embed_query("linkedin account")

results = vector_store.similarity_search_by_vector(queryEmbedding)
print(results[0].page_content)
```
